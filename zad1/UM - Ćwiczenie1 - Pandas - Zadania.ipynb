```python
import pandas as pd
```

```python
#difference

idx1 = pd.Index(['apple', 'banana', 'cherry'])
idx2 = pd.Index(['banana', 'cherry', 'date'])

result = idx1.difference(idx2)
print(result) 
```

```python
#intersection

idx1 = pd.Index([10, 20, 30, 40])
idx2 = pd.Index([30, 40, 50])

common_elements = idx1.intersection(idx2)
print(common_elements)  
```

```python
#union

idx1 = pd.Index(['cat', 'dog'])
idx2 = pd.Index(['dog', 'rabbit'])

result = idx1.union(idx2)
print(result)  
```

```python
#delete

idx = pd.Index(['red', 'blue', 'green'])

result = idx.delete(0)
print(result) 
```

```python
#drop

df = pd.DataFrame({'value': [100, 200]}, index=['A', 'B'])
result = df.drop(index='A')
print(result)
```

```python
#insert

idx = pd.Index([10, 20, 30])

result = idx.insert(1, 29)
print(result) 
```

```python
#is_monotonic

idx_increasing = pd.Index([1, 2, 3])
print(idx_increasing.is_monotonic_increasing)

idx_decreasing = pd.Index([3, 2, 1])
print(idx_decreasing.is_monotonic_decreasing)
 
```

```python
#isin

idx = pd.Index([1, 3, 5, 7])
values = [3, 7, 9]

result = idx.isin(values)
print(result) 
```

```python
#is_unique

idx_unique = pd.Index(['x', 'y', 'z'])
print(idx_unique.is_unique) 

idx_non_unique = pd.Index(['x', 'y', 'x'])
print(idx_non_unique.is_unique) 
```

```python
#unique

idx = pd.Index([5, 5, 10, 15])
result = idx.unique()
print(result) 
```

**Zadanie 3** Podczas wykładu omawiana była metoda `reindex` oraz podane były przykłady jej zastosowania. Zapoznaj się z pozostałymi argumentami metody reindex i wykonaj przykłady z wykorzystaniem tych parametrów.

```python
date_index = pd.date_range('2025-04-01', periods=3, freq='D')
df = pd.DataFrame({'Wartość': [100, 105, 110]}, index=date_index)

new_dates = pd.date_range('2025-03-30', periods=6, freq='D')

#ffill
df_reindexed = df.reindex(new_dates, method='ffill')
print(df_reindexed)
```

```python
new_index = ['A', 'B', 'C', 'D']
df = pd.DataFrame({'X': [1, 2]}, index=['A', 'B'])

df_filled = df.reindex(new_index, fill_value=0)
print(df_filled)
```

```python
s = pd.Series([10, 20, 30], index=[0, 2, 4])

s_reindexed = s.reindex(range(6), method='ffill', limit=1)
print(s_reindexed)
```

```python
idx = pd.Index([10, 20, 30, 40])
target = [9, 15, 25, 45]

new_index, indexer = idx.reindex(target, method='nearest', tolerance=5)
print(new_index)
```

```python
df = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})

df_reindexed = df.reindex(columns=['B', 'A', 'C'], fill_value=0)
print(df_reindexed)
```

```python
df = pd.read_csv('Book_Dataset_1.csv')
```

```python
df = df.rename(columns={"Avilability": "Availability"})
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
cols = df.columns.tolist()
cols.insert(0, cols.pop(cols.index('Category')))
df = df.reindex(columns=cols)
```

```python
df
```

```python
dtypes = df.dtypes

grouped_by_dtype = {}

for col, dtype in dtypes.items():
    grouped_by_dtype.setdefault(dtype, []).append(col)

for dtype, columns in grouped_by_dtype.items():
    print(f"\nTyp danych: {dtype}")
    print("Kolumny:", columns)
```

```python
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df['Availability'] = pd.to_numeric(df['Availability'], errors='coerce')

total_price = (df['Price'] * df['Availability']).sum()

print(f"Total price of all books: {total_price}")
```

```python
df.isnull().any()
```

```python
df.index = df['Title'].str[:3]

duplicate_indices = df.index[df.index.duplicated(keep=False)]
non_unique_counts = df.loc[duplicate_indices].index.value_counts()


print("Wszystkie wartości indeksu, które nie są unikatowe, wraz z liczbą ich wystąpień:")
print(non_unique_counts)
```

```python
import numpy as np
```

```python
tax_dict = {
    category: np.random.randint(0, 21)
    for category in df["Category"].unique()
}

df["Tax_amount"] = df["Category"].map(tax_dict)
```

```python
df
```

```python
df['Price_After_Tax'] = (df['Price'] + df['Tax_amount']).round(2)


print(df[['Price', 'Tax_amount', 'Price_After_Tax']])
```

```python
df.head()
```

```python

history_books = df[(df['Category'] == 'History') & (df['Price_After_Tax'] > 35)]

result_11 = history_books[['Title', 'Number_of_reviews']]
print(result_11)
```

```python
average_price = round(history_books['Price'].mean(), 2)
print(f"Średnia cena książek historycznych: {average_price}")
```

```python
df.loc[history_books.index, 'Price'] = average_price

df.loc[history_books.index, 'Price_After_Tax'] = (
    df.loc[history_books.index, 'Price'] + df.loc[history_books.index, 'Tax_amount']
).round(2)
```

```python
df.head()
```

```python
df['Expected_Tax_Revenue'] = (
    df['Availability'] * df['Price'] * df['Tax_amount']
).round(2)
```

```python
df.head()
```

```python
df = df[~df['Category'].isin(['Horror', 'Mystery'])]
```

```python
filtered_books = df[
    (df['Book_Description'].str.len() > 2000) & 
    (df['Title'].str.len() <= 20)
]

# Wyświetl tytuły
result_16 = filtered_books['Title']
print(result_16)
```
